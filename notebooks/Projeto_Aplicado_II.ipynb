A utilização da Hugging Face Transformers com o modelo RoBERTa é adequada para interpretar reviews de produtos, pois emprega modelos de deep learning capazes de compreender o contexto das frases, incluindo negações e nuances da linguagem natural. O RoBERTa, por ser uma versão otimizada do BERT, apresenta alta precisão na análise de sentimento, permitindo classificar automaticamente os comentários e compará-los com as notas atribuídas pelos usuários de forma eficiente e confiável.

Orientacao do modelo:

Label  - 	Significado

LABEL_0 - 	Negativo

LABEL_1 - 	Neutro

LABEL_2 - 	Positivo

```python
#importando a biblioteca e modelos que serao utilizados
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
from scipy.special import softmax


MODEL = "cardiffnlp/twitter-roberta-base-sentiment"


tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)

classifier = pipeline(
    "sentiment-analysis",
    model=MODEL
)
```

```python
#testes
print(classifier("The product is not good")) #negativa
print(classifier("It is an average item")) #neutra
print(classifier("This product is amazing!")) #positiva
```

```python
model.eval()
```

```python
from tqdm import tqdm
tqdm.pandas()

def calcular_sentimento_roberta(texto):

    try:
        # garantir que é string
        texto = str(texto)

        texto_codificado = tokenizer(
            texto,
            return_tensors='pt',
            truncation=True,
            max_length=512
        )

        saida_modelo = model(**texto_codificado)

        pontuacoes = saida_modelo[0][0].detach().numpy()
        pontuacoes = softmax(pontuacoes)

        negativo = pontuacoes[0]
        neutro = pontuacoes[1]
        positivo = pontuacoes[2]

        score_sentimento = positivo - negativo

        estrelas_previstas = round((score_sentimento + 1) * 2 + 1)
        estrelas_previstas = max(1, min(5, estrelas_previstas))

        return {
            'roberta_negativo': negativo,
            'roberta_neutro': neutro,
            'roberta_positivo': positivo,
            'score_sentimento': score_sentimento,
            'estrelas_previstas': estrelas_previstas
        }

    except Exception as e:
        # debug opcional:
        # print("Erro:", e)

        return {
            'roberta_negativo': None,
            'roberta_neutro': None,
            'roberta_positivo': None,
            'score_sentimento': None,
            'estrelas_previstas': None
        }
```

```python
df_teste = df.head(100)
```

```python
df_resultados = df_teste['Text'].progress_apply(calcular_sentimento_roberta).apply(pd.Series)
```

```python
df_teste = pd.concat([df_teste.reset_index(drop=True), df_resultados], axis=1)
```

```python
df_teste['acerto'] = df_teste['Score'] == df_teste['estrelas_previstas']

df_teste['acerto'].value_counts().plot(
    kind='bar',
    title='Acurácia do Modelo',
)

plt.xticks([0,1], ['Erro', 'Acerto'], rotation=0)
plt.ylabel('Quantidade')

plt.show()
```

Observou-se que o maior erro do modelo ocorreu nas avaliações de nota 2 e 3, o que pode ser explicado pela natureza ambígua desse tipo de feedback. Diferentemente de avaliações claramente positivas ou negativas, as avaliações intermediárias apresentam linguagem mais subjetiva e menos definida, dificultando sua correta classificação. Além disso, o modelo RoBERTa foi treinado para identificar polaridade de sentimento (positivo, neutro e negativo), e não diretamente valores discretos de 1 a 5 estrelas, o que contribui para imprecisões nas notas intermediárias.

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(
    df_teste['Score'],
    df_teste['estrelas_previstas']
)
```

```python
cm_norm = confusion_matrix(
    df_teste['Score'],
    df_teste['estrelas_previstas'],
    normalize='true'
)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm_norm,
    annot=True,
    fmt='.2f',
    cmap='Blues',
    xticklabels=[1,2,3,4,5],
    yticklabels=[1,2,3,4,5]
)

plt.xlabel('Previsto')
plt.ylabel('Real')
plt.title('Matriz de Confusão (Normalizada)')

plt.show()
```

A matriz de confusão evidencia que o modelo apresenta maior precisão nas classificações extremas (notas 1 e 5), com maior concentração de acertos na diagonal principal. Por outro lado, observa-se maior dispersão nas categorias intermediárias, especialmente nas notas 2 e 3, indicando dificuldade do modelo em distinguir avaliações com polaridade menos definida. Esses resultados reforçam a limitação do modelo em mapear corretamente sentimentos ambíguos para uma escala discreta de avaliação.

```python
df_binario = df_teste[df_teste['Score'] != 3].copy()
```

```python
def converter_binario(score):
    if score <= 2:
        return "Negativo"
    else:
        return "Positivo"

df_binario['classe_real'] = df_binario['Score'].apply(converter_binario)
```

```python
def classe_roberta(row):
    if row['roberta_negativo'] >= row['roberta_positivo']:
        return "Negativo"
    else:
        return "Positivo"

df_binario['classe_prevista'] = df_binario.apply(classe_roberta, axis=1)
```

```python
acuracia_binaria = (df_binario['classe_real'] == df_binario['classe_prevista']).mean()

print(f"Acurácia binária: {acuracia_binaria:.2%}")
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(
    df_binario['classe_real'],
    df_binario['classe_prevista'],
    labels=["Negativo", "Positivo"]
)

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=["Negativo", "Positivo"],
    yticklabels=["Negativo", "Positivo"]
)

plt.xlabel("Previsto")
plt.ylabel("Real")
plt.title("Matriz de Confusão (Binário)")

plt.show()
```

```python
df_binario['acerto'] = df_binario['classe_real'] == df_binario['classe_prevista']

df_binario['acerto'].value_counts().reindex([False, True]).plot(kind='bar')

plt.xticks([0,1], ['Erro', 'Acerto'], rotation=0)
plt.title("Acurácia Binária")

plt.show()
```
